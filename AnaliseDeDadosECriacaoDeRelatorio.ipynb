```python
# Importar a base de dados e visualizá-la
# Calcular o faturamento, a quantidade de produtos e o ticket médio de acordo com cada loja
# Criar um relatório para ser encaminhado via e-mail 
```

```python
# Importar as bibliotecas necessárias para manuseio dos dados
import pandas as pd
import win32com.client as win32

# Importar a base de dados
tabela_vendas = pd.read_excel('C:\\Users\\Maria Eduarda\\Downloads\\ProjetosPython\\Vendas.xlsx')
```

```python
# Visualizar a base de dados
pd.set_option('display.max_columns', None)
print(tabela_vendas)
```

```python
# Calcular o faturamento por loja
faturamento = tabela_vendas[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
print(faturamento)

# Calcular a quantidade de produtos vendidos por loja
quantidade = tabela_vendas[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
print(quantidade)

print('-' * 50)
# Calcular o ticket médio por produto em cada loja
ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0: 'Ticket Médio'})
print(ticket_medio)
```

```python
# Enviar um e-mail com o relatório dos dados recolhidos 
# PS: Será necessário instalar o Microsoft Outlook na sua máquina
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)

# Insira aqui o seu endereço de e-mail favorito

mail.To = 'mariaeduardaruck@gmail.com'
mail.Subject = 'Relatório de Vendas por loja'
mail.HTMLBody = f'''
<p>Prezados,</p>

<p>Segue o Relatório de Vendas por cada loja.</p>

<p>Faturamento:</p>
{faturamento.to_html(formatters={'Valor Final': 'R${:,.2f}'.format})}

<p>Quantidade Vendida:</p>
{quantidade.to_html()}

<p>Ticket Médio dos Produtos em cada Loja:</p>
{ticket_medio.to_html(formatters={'Ticket Médio': 'R${:,.2f}'.format})}

<p>Qualquer dúvida estou à disposição.</p>

<p>Att.,</p>
<p>Lira</p>
'''

mail.Send()

print('Email Enviado')
```
